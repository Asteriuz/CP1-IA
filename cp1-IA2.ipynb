#### Importando Bibliotecas Necessárias & Funções de Utilidade

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.datasets import load_diabetes
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
from sklearn.metrics import r2_score, mean_squared_error
from scipy.optimize import minimize
from IPython.display import display, Markdown

def format_df_left(df, coluna_texto='Feature', casas_decimais=3):
    """
    Aplica estilo ao DataFrame: formata as casas decimais, alinha 
    uma coluna de texto à esquerda e oculta o índice.
    """
    estilo = (
        df.style
        .format(precision=casas_decimais)
        .set_properties(subset=[coluna_texto], **{'text-align': 'left'})
        .set_table_styles({
            coluna_texto: [{'selector': 'th', 'props': [('text-align', 'left')]}]
        }, overwrite=False)
        .hide(axis='index')
    )
    
    return estilo
```

#### Pergunta: Os atributos estão em escalas comparáveis? Seria necessário normalizar/re-escalar as features? Por quê?

 As features do dataset Diabetes **já foram padronizadas** pelo scikit-learn
 (média ≈ 0, desvio padrão ≈ 1), portanto estão em escalas comparáveis.

### 3. Modelo inicial (todas as features)

```python
X_all = df[data.feature_names].values
y     = df['target'].values

X_train, X_test, y_train, y_test = train_test_split(
    X_all, y, test_size=0.2, random_state=42)

lr_all = LinearRegression()
lr_all.fit(X_train, y_train)

r2_train_all = r2_score(y_train, lr_all.predict(X_train))
r2_test_all  = r2_score(y_test,  lr_all.predict(X_test))

coef_df_all = pd.DataFrame({
    'Feature': data.feature_names,
    'Coeficiente': lr_all.coef_
})
coef_df_all = coef_df_all.sort_values(by='Coeficiente', ascending=False, key=abs).reset_index(drop=True)
display(format_df_left(coef_df_all, 'Feature'))

metricas_df = pd.DataFrame({
    'Parâmetro': ['Intercepto', 'R² Treino', 'R² Teste'],
    'Valor': [lr_all.intercept_, r2_train_all, r2_test_all]
})
display(format_df_left(metricas_df, 'Parâmetro'))
```

## Parte B - OLS vs ODR com uma única feature (`bmi`)

 ### B.1 Regressão Linear Ordinária (OLS)

```python
feat_escolhida = 'bmi'
X1 = df[feat_escolhida].values
y1 = df['target'].values

X1_tr, X1_te, y1_tr, y1_te = train_test_split(
    X1.reshape(-1, 1), y1, test_size=0.2, random_state=42)

ols = LinearRegression()
ols.fit(X1_tr, y1_tr)

y_pred_ols = ols.predict(X1_te)
rmse_ols   = np.sqrt(mean_squared_error(y1_te, y_pred_ols))
r2_ols     = r2_score(y1_te, y_pred_ols)
m_ols = ols.coef_[0]
b_ols = ols.intercept_

df_ols = pd.DataFrame({
    'Parâmetro': ['Coeficiente (m)', 'Intercepto (b)', 'RMSE Teste', 'R² Teste'],
    'Valor': [m_ols, b_ols, rmse_ols, r2_ols]
})

display(format_df_left(df_ols, 'Parâmetro'))
```

### B.2 Regressão por Distância Ortogonal (ODR / TLS)

OLS minimiza distâncias **verticais** (erro apenas em Y).
ODR minimiza distâncias **perpendiculares** à reta, considerando erro em X e Y.

A distância ortogonal de um ponto $(x_0, y_0)$ até a reta $mx - y + b = 0$ é:

 $$d = \frac{|m x_0 - y_0 + b|}{\sqrt{m^2 + 1}}$$


A função de perda a minimizar é:

 $$\mathcal{L}(m, b) = \sum_i \frac{(m x_i - y_i + b)^2}{m^2 + 1}$$

```python
x_tr = X1_tr.flatten()


def odr_loss(params):
    m, b = params
    return np.sum((m * x_tr - y1_tr + b) ** 2 / (m**2 + 1))


res = minimize(
    odr_loss,
    x0=[m_ols, b_ols],
    method="Nelder-Mead",
    options={"maxiter": 50000, "xatol": 1e-8, "fatol": 1e-8},
)
m_odr, b_odr = res.x

x_te = X1_te.flatten()
y_pred_odr = m_odr * x_te + b_odr
rmse_odr = np.sqrt(mean_squared_error(y1_te, y_pred_odr))
r2_odr = r2_score(y1_te, y_pred_odr)

df_resultados = pd.DataFrame(
    {
        "Parâmetro": [
            "Coeficiente (m)",
            "Intercepto (b)",
            "RMSE Teste",
            "R² Teste",
            "Convergiu",
            "Iterações",
        ],
        "Valor": [m_odr, b_odr, rmse_odr, r2_odr, res.success, res.nit],
    }
)

display(format_df_left(df_resultados, "Parâmetro"))
```

### Comparação OLS vs ODR

```python
x_line = np.linspace(X1.min(), X1.max(), 200)

fig, axes = plt.subplots(1, 2, figsize=(13, 5), sharey=True)
fig.suptitle(f"OLS vs ODR  -  feature: '{feat_escolhida}'", fontsize=13, fontweight='bold')

for ax, (m_, b_, label, color_res) in zip(axes, [
        (m_ols, b_ols, "OLS - distância vertical",   "tomato"),
        (m_odr, b_odr, "ODR - distância ortogonal", "mediumseagreen")]):

    ax.scatter(X1, y1, alpha=0.3, s=18, color='steelblue', label='Dados')
    ax.plot(x_line, m_ * x_line + b_, 'k-', lw=2, label=f'Reta: y={m_:.1f}x+{b_:.1f}')

    xs_ = x_te[:20]; ys_ = y1_te[:20]

    if label.startswith("OLS"):
        for xi, yi in zip(xs_, ys_):
            ax.plot([xi, xi], [yi, m_*xi + b_], color=color_res, lw=0.9, alpha=0.7)
    else:
        for xi, yi in zip(xs_, ys_):
            xp = (xi + m_ * (yi - b_)) / (m_**2 + 1)
            yp = m_ * xp + b_
            ax.plot([xi, xp], [yi, yp], color=color_res, lw=0.9, alpha=0.7)

    ax.set_title(label, fontsize=10)
    ax.set_xlabel(feat_escolhida, fontsize=9)
    ax.set_ylabel("target", fontsize=9)
    ax.legend(fontsize=8)

plt.tight_layout()
plt.show()
```

 ### B.3 Discussão

**1. Em que situações OLS e ODR resultam em retas significativamente diferentes?**

- Quando X possui **erro de medição** relevante (ruído em X "puxa" a reta OLS).
- Na presença de **outliers**: OLS é mais sensível a desvios verticais grandes.
- Quando a razão ruído(X) / ruído(Y) se afasta de zero.

**2. Por que ODR é mais adequado quando tanto X quanto Y têm erro de medição?**

OLS assume que X é medido **sem erro**; toda a incerteza recai sobre Y.
Variáveis biomédicas como IMC e pressão têm imprecisão instrumental.
ODR (Total Least Squares) trata ambas as variáveis como ruidosas,
minimizando a distância perpendicular, produzindo estimativas menos enviesadas.

**3. Como a escala/unidades das variáveis afetam a solução ODR (e por que padronizar é útil)?**

ODR minimiza distâncias euclidianas em (X, Y). Se X estiver em escala muito maior
que Y, o eixo X dominará e a reta será distorcida. Padronizar (z-score) garante
variância ≈ 1 em ambos os eixos, tornando a solução **invariante à escolha de unidades**.

**4. Qual modelo parece generalizar melhor e segundo qual critério?**

| Modelo | RMSE Teste | R² Teste  |
| ------ | ---------- | --------- |
| OLS    | 63.7325    | 0.2334    |
| ODR    | 114.939400 | -1.493500 |

Para **previsão de Y**, OLS é o critério natural (minimiza MSE vertical) e seu RMSE
tende a ser menor. Para **inferência estrutural** entre X e Y com erros em ambos,
ODR é preferível conceitualmente.
